```python
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
print(mpl.__version__)
```

```python
import matplotlib.colors as colors
import numpy as np
```

```python
bounds = np.array([-0.25, -0.125, 0, 0.5, 1])
norm = colors.BoundaryNorm(boundaries=bounds, ncolors=4)
print(norm([-0.2,-0.15,-0.02, 0.3, 0.8, 0.99]))
```

```python
N = 100
X, Y = np.mgrid[-3:3:complex(0, N), -2:2:complex(0, N)]
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

fig, ax = plt.subplots(1, 3, figsize=(24, 10))
ax = ax.flatten()

cmap = 'seismic_r'
# even bounds gives a contour-like effect
bounds = np.linspace(-1, 1, 10)
norm = colors.BoundaryNorm(boundaries=bounds, ncolors=256)
pcm = ax[0].pcolormesh(X, Y, Z,
                       norm=norm,
                       cmap=cmap)
fig.colorbar(pcm, ax=ax[0], extend='both', orientation='horizontal')

# uneven bounds changes the colormapping:
bounds = np.array([-0.25, -0.125, 0, 0.5, 1])
norm = colors.BoundaryNorm(boundaries=bounds, ncolors=256)
pcm = ax[1].pcolormesh(X, Y, Z, norm=norm, cmap=cmap)
fig.colorbar(pcm, ax=ax[1], extend='both', orientation='horizontal')

pcm = ax[2].pcolormesh(X, Y, Z, cmap=cmap, vmin=-np.max(Z))
fig.colorbar(pcm, ax=ax[2], extend='both', orientation='horizontal')
plt.show()
```
