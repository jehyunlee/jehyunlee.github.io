```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# dataset

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

# relational plot

```python
sns.relplot(x="total_bill", y="tip", data=tips, hue="day")
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="day")
```

```python
sns.lineplot(x="total_bill", y="tip", data=tips, hue="day")
```

# distribution plot

> displot
>> histplot  
>> kdeplot  
>> ecdfplot  
>> rugplot  

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species",
            multiple="stack"
           )
```

```python
sns.histplot(data=penguins, x="flipper_length_mm", hue="species",
             multiple="stack"
            )
```

```python
sns.kdeplot(data=penguins, x="flipper_length_mm", hue="species",
            multiple="stack"
           )
```

```python
sns.ecdfplot(data=penguins, x="flipper_length_mm", hue="species")
```

```python
sns.rugplot(data=penguins, x="flipper_length_mm", hue="species")
```
