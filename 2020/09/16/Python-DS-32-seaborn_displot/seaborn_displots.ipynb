```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# data

```python
# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```python
# Load the penguins dataset
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# Load the tips dataset
tips = sns.load_dataset("tips")
tips.head()
```

# set theme

```python
sns.set_theme(style="whitegrid")
```

# distribution

```python
fig, axes = plt.subplots(ncols=3, figsize=(12,4))
sns.histplot(data=diamonds, x="carat", hue="cut", multiple="stack", ax=axes[0], legend=True)
sns.kdeplot(data=diamonds, x="carat", hue="cut", multiple="stack", ax=axes[1], legend=True)
sns.ecdfplot(data=diamonds, x="carat", hue="cut", ax=axes[2], legend=True)

axes[0].set_title("histplot")
axes[1].set_title("kdeplot")
axes[2].set_title("ecdfplot")

fig.tight_layout()
fig.savefig("32_displots_1.png")
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(12,4))
sns.histplot(data=diamonds, x="carat", hue="cut", multiple="fill", ax=axes[0], legend=False)
sns.kdeplot(data=diamonds, x="carat", hue="cut", multiple="fill", ax=axes[1], legend=False)
sns.ecdfplot(data=diamonds, x="carat", hue="cut", ax=axes[2], legend=True)

axes[0].set_title("histplot")
axes[1].set_title("kdeplot")
axes[2].set_title("ecdfplot")

fig.tight_layout()
fig.savefig("32_displots_2.png")
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(8,4))
sns.histplot(data=penguins, x="bill_depth_mm", y="body_mass_g", hue="species", 
             bins=20, ax=axes[0], legend=False)
sns.kdeplot(data=penguins, x="bill_depth_mm", y="body_mass_g", hue="species", 
            fill=True, levels=5, alpha=0.5,
            ax=axes[1], legend=True)

axes[0].set_title("histplot")
axes[1].set_title("kdeplot")

fig.tight_layout()
fig.savefig("32_displots_6.png")
```

```python
# Plot the distribution of clarity ratings, conditional on carat
# https://seaborn.pydata.org/examples/multiple_conditional_kde.html

sns.displot(
    data=diamonds,
    x="carat", hue="cut",
    kind="kde", 
    multiple="fill", clip=(0, None),
    palette="ch:rot=-.25,hue=1,light=.75",
    height=5, aspect=2
)
plt.savefig("32_displots_3.png")
```

```python
# object oriented interface
fig, ax = plt.subplots(figsize=(10, 5))
sns.kdeplot(
    data=diamonds,
    x="carat", hue="cut",
    multiple="fill", clip=(0, None),
    palette="ch:rot=-.25,hue=1,light=.75",
#     height=5, aspect=2
    ax=ax
)
fig.savefig("32_displots_4.png")
```

```python
# object oriented interface
fig, ax = plt.subplots(figsize=(10, 5))
sns.kdeplot(
    data=diamonds,
    x="carat", hue="cut",
    multiple="fill", clip=(0, None),
    palette="ch:rot=-.25,hue=1,light=.75",
#     height=5, aspect=2
    ax=ax
)
h, l = ax.get_legend_handles_labels()
print(f"handles={h}\nlabels={l}\n")
ax.legend(h, l, bbox_to_anchor=(1.05,0.9))
fig.savefig("32_displots_5.png")
```

```python
# object oriented interface
fig, ax = plt.subplots(figsize=(10, 5))
sns.kdeplot(
    data=diamonds,
    x="carat", hue="cut",
    multiple="fill", clip=(0, None),
    palette="ch:rot=-.25,hue=1,light=.75",
#     height=5, aspect=2
    ax=ax
)
ax.legend(diamonds["cut"].unique()[::-1], bbox_to_anchor=(1.2,0.9))
fig.tight_layout()
fig.savefig("32_displots_8.png")
```
