```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_context("talk")
plt.rcParams["font.family"]="NanumGothic"
```

```python
sen = 0.9
spe = 0.99
pre = np.linspace(0.01, 1.0, 100)
```

```python
ppv = (sen * pre)/(sen * pre + (1-spe)*(1-pre)) * 100
npv = (spe * (1-pre))/(spe * (1-pre) + (1-sen)*pre) * 100
```

```python
fig, ax = plt.subplots(figsize=(10, 5), constrained_layout=True)
ax.plot(pre, ppv, c="purple", label="양성예측도 PPV")
ax.plot(pre, npv, c="darkorange", label="음성예측도 NPV")
ax.legend(loc="lower left")
ax.set_xticks(np.linspace(0, 1, 6), list(range(0, 120, 20)))
ax.set_xlim(0, 1)
ax.spines[:].set_visible(False)
ax.spines["bottom"].set_visible(True)
ax.tick_params(axis="y", length=0)
ax.grid(axis="y", c="lightgray")
ax.axvline(0.326, c="cornflowerblue")
ax.text(0.326, 1.02, "32.6% (2022.04.23.)", transform=ax.transAxes, ha="center",
       c="cornflowerblue")
ax.set_xlabel("유병률 (%)")

fig.savefig("99_cp_07.PNG")
```

```python
fig, ax = plt.subplots(figsize=(10, 5), constrained_layout=True)

spe = 0.99
sens = [0.9, 0.8, 0.7, 0.6, 0.5]
pal_ppv = sns.color_palette("light:purple_r", 6)
pal_npv = sns.color_palette("light:darkorange_r", 6)

for sen, c_ppv, c_npv in zip(sens, pal_ppv, pal_npv):
    ppv = (sen * pre)/(sen * pre + (1-spe)*(1-pre)) * 100
    npv = (spe * (1-pre))/(spe * (1-pre) + (1-sen)*pre) * 100
    
    ax.plot(pre, ppv, c=c_ppv, label=f"민감도={int(sen*100)}%")
    ax.plot(pre, npv, c=c_npv, label=f"민감도={int(sen*100)}%")

ax.set_xticks(np.linspace(0, 1, 6), list(range(0, 120, 20)))
ax.set_xlim(0, 1)
ax.spines[:].set_visible(False)
ax.spines["bottom"].set_visible(True)
ax.tick_params(axis="y", length=0)
ax.grid(axis="y", c="lightgray")
ax.axvline(0.326, c="cornflowerblue")

ax.set_xlabel("유병률 (%)")

handles, legend = ax.get_legend_handles_labels()
h_ppvs = [h for i, h in enumerate(handles) if i%2 == 0]
h_npvs = [h for i, h in enumerate(handles) if i%2 == 1]
ax.legend(loc="upper left", ncol=1, handles=h_ppvs + h_npvs,
         title="PPV(보라), NPV(주황)",
         bbox_to_anchor=(1, 1))
ax.set_title("특이도 = 99%", fontweight="bold")

fig.savefig("99_cp_08.PNG")
```
