```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv("data1.csv")
data1.head()
```

```python
data2 = pd.read_csv("data2.csv")
data2.head()
```

## 2D

```python
fig, axs = plt.subplots(ncols=2, figsize=(8, 4), constrained_layout=True)

fontlabel = {"fontsize":"large", "color":"gray", "fontweight":"bold"}
for ax, data in zip(axs, [data1, data2]):
    ydata = "Y1" if data is data1 else "Y2"
    ax.scatter(data["X"], data[ydata], c=data["Z"], cmap="inferno", s=10)
    ax.set_xlabel("X", fontdict=fontlabel)
    ax.set_ylabel(ydata, fontdict=fontlabel)
    
fig.savefig("mpl3d_03.png")
```

```python
# imshow

fig, axs = plt.subplots(ncols=2, figsize=(8, 4), constrained_layout=True)

fontlabel = {"fontsize":"large", "color":"gray", "fontweight":"bold"}
for ax, data in zip(axs, [data1, data2]):
    ydata = "Y1" if data is data1 else "Y2"
    ax.imshow(data.pivot_table("Z", ydata, "X"), cmap="inferno", 
              extent=[data["X"].min(), data["X"].max(), data[ydata].max(), data[ydata].min()])
    ax.invert_yaxis()
    ax.set_aspect((data["X"].max() - data["X"].min())/(data[ydata].max() - data[ydata].min()))
    ax.set_xlabel("X", fontdict=fontlabel)
    ax.set_ylabel(ydata, fontdict=fontlabel)
    
fig.savefig("mpl3d_05.png")
```

```python
# pcolor

fig, axs = plt.subplots(ncols=2, figsize=(8, 4), constrained_layout=True)

fontlabel = {"fontsize":"large", "color":"gray", "fontweight":"bold"}
for ax, data in zip(axs, [data1, data2]):
    ydata = "Y1" if data is data1 else "Y2"
    data_pt = data.pivot_table("Z", ydata, "X")
    ax.pcolor(data_pt, cmap="inferno")
    ax.set_xlabel("X", fontdict=fontlabel)
    ax.set_ylabel(ydata, fontdict=fontlabel)
    
    xticks = [x for x in ax.get_xticks() 
              if (ax.get_xbound()[0] <= x <= ax.get_xbound()[1]) 
                  and x < len(data_pt.columns)]
    ax.set_xticks(xticks)
    ax.set_xticklabels([f"{data_pt.columns[int(x)]:.3f}" for x in xticks])
    
    yticks = [y for y in ax.get_yticks() 
              if (ax.get_ybound()[0] <= y <= ax.get_ybound()[1]) 
                  and y < len(data_pt.index)]
    ax.set_yticks(yticks)
    ax.set_yticklabels([f"{data_pt.index[int(y)]:.3f}" for y in yticks])
    
fig.savefig("mpl3d_06_1.png")
```

```python
from matplotlib import animation 

fig, axs = plt.subplots(ncols=2, figsize=(10, 5), subplot_kw={"projection":"3d"})

fontlabel = {"fontsize":"large", "color":"gray", "fontweight":"bold"}

def init():
    for ax, data in zip(axs, [data1, data2]):
        ydata = "Y1" if ax == axs[0] else "Y2"
        ax.set_xlabel("X", fontdict=fontlabel, labelpad=16)
        ax.set_ylabel(ydata, fontdict=fontlabel, labelpad=16)
        ax.set_title("Z", fontdict=fontlabel)
    
        ax.scatter(data["X"], data[ydata], data["Z"], 
                   c=data["Z"], cmap="inferno", s=5, alpha=0.5)
    
    return fig,

def animate(i):
    axs[0].view_init(elev=30., azim=i)
    axs[1].view_init(elev=30., azim=i)
    return fig,

# Animate
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=360, interval=20, blit=True)
# Save
anim.save('mpl3d_scatter.gif', fps=30)
```
