```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_fs = pd.read_excel("frictionstress.xlsx")
df_fs
```

```python
# 합금 이름
alloy = df_fs.iloc[:, 0].values.copy()

# elastic modulus
ela = df_fs.iloc[:, 1].values.copy()

# friction stress
sigma = df_fs.iloc[:, 2].values.copy()

# normalized friction stress
sigma_norm = sigma / ela
```

```python
# 시각화 설정
import seaborn as sns
sns.set_context("talk")
sns.set_style("white")

# 시각화
fig, ax = plt.subplots(figsize=(6, 6))
ax.barh(alloy, sigma_norm)

fig.tight_layout()
fig.savefig("89_matprop_02.png")
plt.show()
```

```python
# 정렬
df_fs["sigma_norm"] = sigma_norm
df_fs = df_fs.sort_values("sigma_norm")
df_fs
```

```python
# 합금 이름
alloy = df_fs.iloc[:, 0].values.copy()
sigma_norm = df_fs.iloc[:, 3].values.copy()
```

```python
# 시각화
fig, ax = plt.subplots(figsize=(6, 6))
ax.barh(alloy, sigma_norm)

fig.tight_layout()
fig.savefig("89_matprop_04.png")
plt.show()
```

```python
# 그래프 꾸미기
fig, ax = plt.subplots(figsize=(6, 6))
ax.barh(alloy, sigma_norm, fc="lightgray")

# 막대들 모으기
bars = ax.patches
bars[-3].set_facecolor("yellow")
bars[-3].set_edgecolor("k")
bars[-3].set_linewidth(1)

fig.tight_layout()
fig.savefig("89_matprop_05.png")
plt.show()
```

```python
# 0.6 넘는 물질들 색칠
fig, ax = plt.subplots(figsize=(6, 6))
ax.barh(alloy, sigma_norm, fc="lightgray")

# 막대들 모으기
bars = ax.patches

for bar in bars:
    if bar.get_width() > 0.6:
        bar.set_facecolor("orange")
        bar.set_linewidth(1)

bars[-3].set_facecolor("yellow")
bars[-3].set_edgecolor("k")
bars[-3].set_linewidth(1)

fig.tight_layout()
fig.savefig("89_matprop_18.png")
plt.show()
```

```python
# 0.6 넘는 물질들 색칠
fig, ax = plt.subplots(figsize=(6, 6))
ax.barh(alloy, sigma_norm, fc="lightgray")

# 막대들 모으기
bars = ax.patches

for bar in bars:
    if bar.get_width() > 0.6:
        bar.set_facecolor("orange")
        bar.set_linewidth(1)

bars[-3].set_facecolor("yellow")
bars[-3].set_edgecolor("k")
bars[-3].set_linewidth(1)

ax.set_title("Friction Stress ($\sigma_0 / E$, 10$^{-3}$)", pad=20)

fig.tight_layout()
fig.savefig("89_matprop_19.png")
plt.show()
```

```python
# 0.6 넘는 물질들 색칠
fig, ax = plt.subplots(figsize=(6, 6))
ax.barh(alloy, sigma_norm, fc="lightgray")

bars = ax.patches

for bar in bars:
    if bar.get_width() > 0.6:
        bar.set_facecolor("orange")
        bar.set_linewidth(1)

bars[-3].set_facecolor("yellow")
bars[-3].set_edgecolor("k")
bars[-3].set_linewidth(1)

ax.set_title("Friction Stress ($\sigma_0 / E$, 10$^{-3}$)", pad=20)

# x축 눈금 지정
xticks = np.linspace(0, 1.2, 5)
ax.set_xticks(xticks)

fig.tight_layout()
fig.savefig("89_matprop_20.png")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.barh(alloy, sigma_norm, fc="lightgray")

bars = ax.patches
for bar in bars:
    if bar.get_width() > 0.6:
        bar.set_facecolor("orange")
        bar.set_linewidth(1)

bars[-3].set_facecolor("yellow")
bars[-3].set_edgecolor("k")
bars[-3].set_linewidth(1)

ax.set_title("Friction Stress ($\sigma_0 / E$, 10$^{-3}$)", pad=20)

xticks = np.linspace(0, 1.2, 5)
ax.set_xticks(xticks)

# 테두리 지우기
ax.spines[["top", "right", "bottom"]].set_visible(False)

# 눈금 넣기
ax.grid(axis="x")

fig.tight_layout()
fig.savefig("89_matprop_21.png")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.barh(alloy, sigma_norm, fc="lightgray")

bars = ax.patches
for bar in bars:
    if bar.get_width() > 0.6:
        bar.set_facecolor("orange")
        bar.set_linewidth(1)

bars[-3].set_facecolor("yellow")
bars[-3].set_edgecolor("k")
bars[-3].set_linewidth(1)

ax.set_title("Friction Stress ($\sigma_0 / E$, 10$^{-3}$)", pad=20)

xticks = np.linspace(0, 1.2, 5)
ax.set_xticks(xticks)

ax.spines[["top", "right", "bottom"]].set_visible(False)
ax.grid(axis="x")

# alloy 이름에 아래첨자 추가
yticks = ax.get_yticks()
yticklabels = alloy
yticklabels[1] = "(CoNi)$_{95}$Cr$_5$"
yticklabels[-3] = "(CoNi)$_{80}$Cr$_{20}$"
yticklabels[-1] = "Co$_{20}$(CrNi)$_{80}$"
ax.set_yticks(yticks)
ax.set_yticklabels(yticklabels)

fig.tight_layout()
fig.savefig("89_matprop_22.png")
plt.show()
```

```python
yticks = ax.get_yticks()
yticklabels = ax.get_yticklabels()

# 물질 이름 색상, 크기 수정
for bar, ytl in zip(bars, yticklabels):
    if bar.get_width() > 0.6:
        ytl.set_color("peru")
    ytl.set_fontsize("small")
        
yticklabels[-3].set_fontweight("bold")
yticklabels[-3].set_fontsize("medium")

fig.tight_layout()
fig.savefig("89_matprop_23.png")
display(fig)
```
