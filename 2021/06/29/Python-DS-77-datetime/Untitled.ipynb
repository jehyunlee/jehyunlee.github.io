```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = np.random.random((5, 5))
X
```

```python
ax = sns.heatmap(X)
```

```python
ax.get_children()
```

```python
fig = ax.figure
fig.get_children()
```

```python
print(fig.get_children()[2].get_yticks())
print(fig.get_children()[2].get_yticklabels())
```

```python
yticks_new = [0, 0.5, 1]
fig = ax.figure
fig.get_children()[-1].yaxis.set_ticks(yticks_new)

display(fig)
```
