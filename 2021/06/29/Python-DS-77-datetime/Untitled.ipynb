```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.linspace(0, 100, 1001)
x2 = np.linspace(100, 110, 101)

y1 = np.sin(x1)
y2 = np.sin(x2)

fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(x1, y1, ".-", ms=5, alpha=0.3)
ax.plot(x2, y2, ".-", ms=5, alpha=0.3)
plt.show()
```

```python
ax.get_children()
```

```python
ax.get_children()[0].set_color("r")
ax.get_children()[0].set_markeredgecolor("r")
ax.get_children()[0].set_markerfacecolor("r")
```
